## Задание 1

Постройте линейную диаграмму, отображающую температуру в городе за одну неделю. Создайте два списка чисел, один для оси X, представляющий дни недели, а другой для оси Y, представляющий температуру. Добавьте подписи к осям и заголовок графика.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

x = np.random.randint(20,35,size=7)
y = pd.Series([1,2,3,4,5,6,7])

plt.figure(figsize=(10,6))
plt.plot(y,x,c='blue',marker='*')
plt.xlabel('Day of the Week')
plt.ylabel('Temperature of the Week')
plt.title('')
plt.show()
```

## Задание 2

Постройте точечную диаграмму, иллюстрирующую взаимосвязь между количеством часов учёбы и оценками студентов. Создайте два списка чисел, один для оси X, представляющий часы учебы, а другой для оси Y, представляющий оценки студентов. Добавьте подписи к осям и заголовок.



```python
np.random.seed(42)
spend_hours = np.random.rand(10) * 10
marks = np.random.randint(30,100,size=10)

plt.figure(figsize=(10,6))
plt.scatter(spend_hours,marks,c='g',marker ='*')
plt.xlabel('Часы Учебы')
plt.ylabel('Оценки Студентов')
plt.title('Корреляция Между Часами для Учебы и Оценки студентов')
plt.show()
```

## Задание 3

Создайте список категорий (например, разные виды спорта) и список значений (например, количество участников в каждом виде спорта). Постройте столбчатую диаграмму для сравнения количества участников. Добавьте подписи к осям и заголовок.



```python
np.random.seed(42)
sport_name = ['Basketball','Box', 'Judo', 'Karate', 'Grek-rim wrestling', 'Volleyball', 'Soft Athlethic']
participeted = np.random.randint(50,200,size=len(sport_name))

plt.figure(figsize=(10,6))
plt.bar(sport_name, participeted)
plt.xlabel('Вид Спорта', c = 'r')
plt.ylabel('Кол-Во Участников')
plt.show()
```

## Задание 4

Создайте гистограмму для анализа частоты встречаемости различных оценок. Создайте набор данных, используя генерацию случайных чисел. Например, оценки могут быть в диапазоне от 2 до 5 (включительно). Для этого используйте библиотеку numpy (функцию numpy.random.randint) Выберите подходящее количество бинов для гистограммы. Например, для оценок может подойти 8 бина. Добавьте подписи к осям и заголовок.



```python
np.random.seed(42)

grades = np.random.randint(2,6,size=100)

plt.figure(figsize=(10, 6))
plt.hist(grades, bins=4, edgecolor='black', color='skyblue')

# Добавление заголовка и подписей осей
plt.title('Распределение оценок', fontsize=16)
plt.xlabel('Оценки', fontsize=14)
plt.ylabel('Количество', fontsize=14)

# Включение сетки
plt.grid(True)

# Отображение графика
plt.show()
```

## Задание 5

Создайте две Столбчатые Диаграммы на одном изображении:

Первая Диаграмма: Постройте столбчатую диаграмму, показывающую количество чашек разных видов кофе (например, эспрессо, латте, капучино), проданных в кофейне за день.
Вторая Диаграмма: Создайте столбчатую диаграмму для пяти различных закусок (например, круассаны, багеты, печенья и т.д.), также проданных в этот же день.


```python
np.random.seed(42)

coffee_type = ['Espresso', 'Latte', 'Capuchino','Americano', 'Cortado']
coffee_solded = np.random.randint(5,101,size=5)

delishes = ['Cruassan', 'Baget', 'Cookie', 'Cakes', 'Piroschki']
delishes_solded = np.random.randint(5,101,size=5)


plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.bar(coffee_type, coffee_solded,edgecolor='black',color='tomato',linewidth=2)
plt.xlabel('Тип Кофе',c='r',fontsize=10)
plt.ylabel('Кол-во Продаж',fontsize=10)

plt.subplot(1,2,2)
plt.bar(delishes, delishes_solded,edgecolor='black',color='skyblue', linewidth=2)
plt.xlabel('Вид Закуски',color = 'r',fontsize=10)
plt.ylabel('Кол-во Продаж',fontsize=10)
plt.tight_layout()
plt.show()
```
