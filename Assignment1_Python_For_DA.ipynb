```python
# Filtering paratelEducation =3 -> Bachaleour
data = data[data['ParentalEducation']==3].reset_index()
data.tail()
```

##  Task 1

```python
# Processing our data into small segments

# Grouping our data by age and count their quantity
age = data.copy()
age = age.groupby('Age').size().reset_index(name='Count')
age

# 
ethnicity = data.copy()
ethnicity = ethnicity.groupby('Ethnicity').size().reset_index(name='Count')
ethnicity
ethnicity['Name of Ethnicity'] = pd.DataFrame({'Name of Ethnicity':['Caucasian','African American','Asian','Other']})


grade = data.copy()
grade = grade.groupby('GradeClass').size().reset_index(name='Count')
grade['GradeClass']= grade['GradeClass'].astype(int)
```

```python
# Visualization part with bar Chart or Pie Chart or Scatter plot
# Plot a pie chart instead

plt.figure(figsize=(4,4))
plt.title('Students Age in Bar Chart')
plt.bar(age['Age'], age['Count'],edgecolor='black',color='skyblue')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization Ethnicity 

plt.bar(ethnicity['Name of Ethnicity'],ethnicity['Count'],color='tomato', edgecolor='black')
plt.title('Difference Ethnicity Distribution')
plt.show()
```

```python
# Grade distribution 

grade['GradeClass'] = grade['GradeClass'].replace({0:'A',1:'B',2:'C',3:'D',4:'F'})

# Visualisation part
plt.bar(grade['GradeClass'], grade['Count'], color= ['yellow','gold','orange','red','darkred'],edgecolor='black')
plt.title('Grade Distribution')
plt.show()
```

## Key insights from Task 1
  1) The Students whom have F is A LOT. Approx -> 200 

## Task 2 Pie Charts and Boxplot -> {gender, age}

```python
# Distibute gender by male and female
gender = data.copy()
gender = gender.groupby('Gender').size().reset_index(name='Count')
gender['Gender Name'] = pd.DataFrame({'Gender Name':['Male','Female']})

distr_age_by_gender = data.copy()
distr_age_by_gender['Gender'] = distr_age_by_gender['Gender'].replace({1:'Female',0:'Male'})
```

```python
# Data Visualization
plt.figure(figsize=(12,6)) # Our figure size

plt.subplot(1,2,1) # Using Subplot for input more than one visualization
plt.pie(gender['Count'],labels=gender['Gender Name'],autopct='%1.1f%%', colors=['royalblue','deeppink'])
plt.legend(gender['Gender Name'])
plt.title('Pie chart of Destribution by Students Gender',fontsize=10)

# Creating second vizalisation
plt.subplot(1, 2, 2) # It will be second visualization
# Creating the boxplot with using seaborn
sns.boxplot(x='Gender', y='Age', data=distr_age_by_gender, hue='Gender', palette=['royalblue', 'deeppink'], legend=True) 
plt.title('Boxplot of Age by Gender')
plt.show()
```

## Key insights from Task 2
 1. Girls Dominating over Boys
 2. Mean Distibustion age of Male students -> [15 and 17]
 3. Mean Distibution age of Female students -> [16 and 18]

## Task 3 Corralation between Student's GPA and Their absence

```python
# Creating and filtering our data
colleration_gpa_and_grades = data.copy()
colleration_gpa_and_grades['GradeClass'] = colleration_gpa_and_grades['GradeClass'].replace({0:'A',1:'B',2:'C',3:'D',4:'F'})
colleration_gpa_and_grades = colleration_gpa_and_grades.sort_values('GradeClass')

# Creating boxlplot for corralation
# Visualization
sns.boxplot(x='GradeClass',y='Absences',data=colleration_gpa_and_grades,hue='GradeClass',palette=['yellow','gold','orange','red','darkred'],legend=False)
plt.title("Corroletion of Students's GPA and Absence")
plt.show()
```

## Key insights from Task 3
 1. Students in the lower grades (especially class 0) have higher grades (GPAs) than students in the upper grades.
 2. Classes 1, 2, and 3 have more stable GPAs, but there are significant outliers.

 3. Class 4 has the lowest average GPAs and a large variance in results, which may indicate that students in this class are struggling academically.

## Task 4 

```python
# Preparing our data 
task4 = data.copy()
task4['ParentalSupport'] = task4['ParentalSupport'].replace({0:'Not',
                                          1:'Low',
                                          2:'Modarete',
                                          3:'High ',
                                          4:'Very High'})

# Creating our figure with it's size 
plt.figure(figsize=(13,5))
plt.subplot(1,2,1) # It'll be our first vizualization
sns.boxplot(x='ParentalSupport',y='Absences',data=task4,hue='ParentalSupport', legend=False)

# It'll be our second visualization
plt.subplot(1,2,2)
sns.boxplot(x='ParentalSupport',y='GPA',data=task4,hue='ParentalSupport',legend=False)
plt.show()
```

## Key insights from Task 4
 1. Parent Support in very important and it has very influence for students behavior
 2. Absence and Gpa students whom have a not support have bad marks
