```python
import pandas as pd 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import datasets


wine_data = datasets.load_wine()
wine_df = pd.DataFrame(wine_data.data,columns=wine_data.feature_names)
wine_df['wine_type'] = wine_data.target
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.scatterplot(data=wine_df,x = 'alcohol', y = 'magnesium', hue='wine_type', palette='viridis',marker='x')
plt.ylabel('Magnesium Level')
plt.xlabel('Alcohol Content')
plt.title('Scatter PLot of Alchohol Content vs Magnesium Level in Wine by Wine Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=wine_df, x="wine_type", y="alcohol" ,hue='wine_type', palette='viridis')
plt.title("Box Plot of Alcohol Content Distribution by Wine Type")
plt.xlabel("Wine Type")
plt.ylabel("Alcohol Content")
plt.show()
```

```python
# Training model

X = wine_df.drop('wine_type',axis=1)
y = wine_df['wine_type']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model = LogisticRegression(max_iter=10000)
model.fit(X_train,y_train)
```

```python
y_prediction = model.predict(X_test)
acuracy = accuracy_score(y_test,y_prediction)
print('Accruracy of this model is -> {:.2f}'.format(acuracy))
```
