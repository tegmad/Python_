__7. Filter the datasets to only include athletes who won at least two medals. Subset the data further to focus on the last 20 years of both Summer and Winter Olympics.__

__8. Create a new dataset summarizing the total number of medals won by each country, by sport and season (Summer/Winter), using multiple aggregation functions (e.g., total medals, average medals per athlete).__

```python
summer = dataSummer.copy()
summer['Olimpic Time']= 'Summer'
summer['Coefficient'] = summer['Medal'].map({'Gold':3,'Silver':2, 'Bronze':3})
summer['Country'] = summer['Country'].fillna('UK')
summer.isnull().sum()

winter = dataWinter.copy()
winter ['Olimpic Time'] = 'Winter'
winter['Coefficient'] = winter['Medal'].map({'Gold':3,'Silver':2, 'Bronze':3})
winter.isnull().sum()
total = pd.concat([summer, winter],axis=0)
total = total.drop_duplicates()

total1 = total.copy() 
total1 = total.groupby(['Year','Olimpic Time','Country'])['Coefficient'].agg(['count','sum','mean']).sort_values('sum', ascending=False).head(10)
total1 = total1.rename(columns={'count':'Medal Counts',
                                'sum':'Coefficient Sum',
                                'mean':'Mean'})
total1
```

__9. Create a wide-format dataset where each row represents a country, and the columns show the total number of gold, silver, and bronze medals won by year for both Summer and Winter Olympics.__

```python
wide_dataFrame = total.copy()

medal_counts = wide_dataFrame.groupby(['Year', 'Olimpic Time', 'Country', 'Medal']).size().unstack(fill_value=0)
medal_counts['Total Medals'] = medal_counts.sum(axis=1)
medal_counts
```
